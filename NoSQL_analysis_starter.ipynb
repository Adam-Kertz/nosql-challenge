## Notebook Set Up

```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
hygiene_query = {'scores.Hygiene': 20}
hygiene_results = establishments.find(hygiene_query)

# Use count_documents to display the number of documents in the result
print(f"There are {establishments.count_documents(hygiene_query)} establishments with a hygiene score of 20.\n")

# Display the first document in the results using pprint
print("Here is an example:")
pprint(hygiene_results[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_results = establishments.find(hygiene_query)
hygiene_df = pd.DataFrame(hygiene_results).set_index('_id',drop=True)

# Display the number of rows in the DataFrame
print(f"There are {hygiene_df.shape[0]} rows in the DataFrame.")

# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
London_query = {'LocalAuthorityName':{'$regex':'London'}}
rating_query = {'$and':[London_query,{'RatingValue': {'$gte': 4}}]}
rating_results = establishments.find(rating_query)

# Use count_documents to display the number of documents in the result
print(f"There are {establishments.count_documents(rating_query)} London establishments with a rating greater than or equal to 4.\n")

# Display the first document in the results using pprint
print("Here is an example:")
pprint(rating_results[0])
```

```python
# Convert the result to a Pandas DataFrame
rating_results = establishments.find(rating_query)
rating_df = pd.DataFrame(rating_results).set_index('_id',drop=True)

# Display the number of rows in the DataFrame
print(f"There are {rating_df.shape[0]} rows in the DataFrame.")

# Display the first 10 rows of the DataFrame
rating_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

# Define degree of variance
degree_search = 0.01

# Get information on Penang Flavours Establishment
new_business_query = {"BusinessName":"Penang Flavours"}
new_business_results = establishments.find_one(new_business_query)

# Obtain latitude and longitude from query
latitude = new_business_results.get('geocode').get('latitude')
longitude = new_business_results.get('geocode').get('longitude')

# Define latitude and longitude limits for new query
lat_min = latitude - degree_search
lat_max = latitude + degree_search
lon_min = longitude - degree_search
lon_max = longitude + degree_search

# Define query and sort criteria for restaurants in desired range
distance_query = {'$and':[{'geocode.latitude':{'$gte': lat_min}},{'geocode.latitude':{'$lte':lat_max}}
                 ,{'geocode.longitude':{'$gte':lon_min}},{'geocode.longitude':{'$lte':lon_max}}]}
distance_sort = [('scores.Hygiene', 1)]

# Find establishments of note
distance_results = establishments.find(distance_query).sort(distance_sort).limit(5)

# Print the results
for result in distance_results:
    pprint(result)
```

```python
# Convert result to Pandas DataFrame
distance_results = establishments.find(distance_query).sort(distance_sort).limit(5)
distance_df = pd.DataFrame(distance_results).set_index('_id',drop=True)
distance_df.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
pipeline_match = {'$match':{'scores.Hygiene': 0} }
# 2. Groups the matches by Local Authority
pipeline_group = {'$group': {'_id': "$LocalAuthorityName", 'count': {'$sum' : 1}}}
# 3. Sorts the matches from highest to lowest
pipeline_sort = {'$sort': { 'count': -1 }}

# Define pipeline and run aggregate query
pipeline = [pipeline_match, pipeline_group, pipeline_sort]
results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f"There are {len(results)} documents in the result.")

# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
results_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(f"There are {results_df.shape[0]} rows in the results dataframe")

# Display the first 10 rows of the DataFrame
results_df.head(10)
```
